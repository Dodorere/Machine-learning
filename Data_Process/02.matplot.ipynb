```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
xarr = np.random.rand(15).cumsum()
xarr
yarr = np.random.rand(15).cumsum()
yarr
```

```python
plt.figure(figsize=(10,3))
plt.plot(xarr)
plt.show()
#1차원 데이터의 경우 데이터 값을 y축으로, 인덱스 값을 x축으로 하여 그려준다.
```

```python
plt.figure() #만들어도 보여지는 것은 없음, 사이즈 바꿀 때 필요
plt.subplot() #axes 생성
plt.plot(xarr)
plt.show()
#눈에 보여주는 것은 axes 영역
```

```python
# 플롯을 여러개 그리는 방법 1
```

```python
plt.figure() 
plt.subplot(1,2,1) 
plt.plot(xarr)
plt.subplot(1,2,2) 
plt.plot(xarr)
plt.plot(yarr)
plt.show()
```

```python
# 플롯을 여러개 그리는 방법 2
```

```python
fig = plt.figure() 
ax1 = fig.add_subplot(1,2,1) 
ax1.plot(xarr)
ax2 = fig.add_subplot(1,2,2) 
ax2.plot(yarr)
plt.show()
```

```python
# 플롯을 여러개 그리는 방법 3
```

```python
fig, axes = plt.subplots(2,2)
axes[0,0].plot(xarr)
plt.show()
```

```python
fig, axes = plt.subplots(2,2)
axes = axes.flatten()

for i in range(4):
    axes[i].plot(xarr)
plt.show()
#loop 돌리기
```

```python
xarr = np.random.rand(15).cumsum()
yarr = np.random.rand(15).cumsum()
zarr = np.random.rand(15).cumsum()
```

```python
plt.figure(figsize=(8,6))
plt.plot(xarr, label="Korean", color="#FB1495", linestyle="--", marker="o")
plt.plot(yarr, label="English",color="#119944", linestyle="-.", marker="^")
plt.plot(zarr, label="Japanese",color="#7799FF", linestyle=":", marker="*")
plt.title("This is title")
plt.xlabel("X lable")
plt.ylabel("Y lable")
plt.legend()
plt.show()
```

## Scatter plot(산점도)

```python
plt.scatter(xarr,yarr,color="#21F9F4", s=50, marker="^")
plt.scatter(xarr,zarr)
plt.show()
```

## Bar plot

```python
year = ["A", "B", "C"]
points = [100, 200, 300]

plt.bar(year, points, color="#1F970F", width=0.3)
plt.show()
```

```python
plt.barh(year, points)
plt.show()
```

## Box plot

```python
x = np.random.normal(0,2.0,100)
y = np.random.normal(-3,1.5,500)
z = np.random.normal(1.2,1.5,1000)
```

```python
plt.boxplot([x,y,z])
plt.show()
#중간의 노란색선은 median(중간값, q2라고 보면 됨)
#박스의 길이가 길면 데이터가 그만큼 퍼져있다는 뜻
#박스 밖의 원들은 이상한 데이터 값들
```
