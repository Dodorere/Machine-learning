```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data = pd.read_csv('datasets/activity_train.csv')
```

```python
data.head(10)
```

```python
df = data.copy()
```

```python
df = df.drop(['subject'], axis = 1)
```

```python
df.head()
```

```python
x_train = df.iloc[:, :561]
```

```python
y_train = df.iloc[:, 561]
y_train = y_train.map({"STANDING":0, "SITTING":1, "LAYING":2, "WALKING":3, "WALKING_DOWNSTAIRS":4, "WALKING_UPSTAIRS":5})
```

```python
y_train.value_counts()
```

```python
x_test = df.iloc[:, :561]
y_test = df.iloc[:, 561]
```

```python
y_test = y_test.map({"STANDING":0, "SITTING":1, "LAYING":2, "WALKING":3, "WALKING_DOWNSTAIRS":4, "WALKING_UPSTAIRS":5})
```

```python
## KNN

from sklearn.neighbors import KNeighborsClassifier
from sklearn.model_selection import cross_val_score
from tqdm import tqdm
#tqdm으로 모델이 돌아가고 있는지 확인 가능

hist = []

for k in tqdm(range(3,50)):
    model = KNeighborsClassifier(n_neighbors = k)
    scores = cross_val_score(model, x_train, y_train, cv = 5, scoring = 'accuracy')
    
    hist.append(scores.mean())
```

```python
np.argmax(hist)
```

```python
model = KNeighborsClassifier(n_neighbors = 14)
model.fit(x_train, y_train)
y_pred = model.predict(x_test)
```

```python
from sklearn.metrics import accuracy_score

accuracy_score(y_test, y_pred)
```

```python
##SVM - linear
```

```python
from sklearn import svm

clf = svm.SVC(kernel = 'linear')
clf.fit(x_train, y_train)
```

```python
y_pred = clf.predict(x_test)
accuracy_score(y_test, y_pred)
```

```python
from sklearn.metrics import confusion_matrix

#앞에 정답값을 넣어야함
cfm = confusion_matrix(y_test, y_pred)
cfm

plt.figure(figsize = (4,4))
sns.heatmap(cfm, annot = True)
plt.show()
```

```python
## DT
```

```python
from sklearn.tree import DecisionTreeClassifier

dt = DecisionTreeClassifier()
dt.fit(x_train, y_train)
```

```python
y_pred = dt.predict(x_test)
accuracy_score(y_test, y_pred)
```

```python
## DT-gridsearch
```

```python
from sklearn.model_selection import GridSearchCV

param = {
    'max_depth' : range(1, 10),
    'min_samples_split' : [2,3,4],
    'min_samples_leaf' : range(1,10)
}

model = DecisionTreeClassifier()
grid = GridSearchCV(model, param, refit = True)
grid.fit(x_train, y_train)
```

```python
grid.best_params_
```

```python
y_pred = grid.best_estimator_.predict(x_test)
```

```python
accuracy_score(y_test, y_pred)
```
